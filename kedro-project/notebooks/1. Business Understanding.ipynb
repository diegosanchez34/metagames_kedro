# Phase 1: Business Understanding

MetaGames is an online gaming platform developed by a Chilean company with the aim of offering a local alternative to international giants such as Steam and Epic. The platform seeks to stand out through more accessible prices and payment systems adapted to the needs of the Chilean market. Currently, MetaGames provides users with an extensive catalog of PC games organized into various categories, as well as tables that highlight the games with the best ratings and the most played. With the continued growth of the platform, the company is focused on improving the user experience through the implementation of new features, including the creation of a personalized recommendation system.

## Main Objectives:

### Developing a Personalized Recommendation System

- Target for the regression task: In the context of this project, the goal for the regression task is to predict a game's affinity score for a user. This score will reflect how relevant a new game might be to the user based on their purchase history and the games in their library. This target was chosen because it provides a quantitative value that can be used to rank recommendations. This prediction allows the system to fine-tune suggestions, which directly impacts the business goal of improving user experience, as higher affinity increases the likelihood of engagement and purchasing recommended games. Selecting this target is key to making subsequent decisions, such as optimizing the recommendation algorithm and personalizing offers.

- Target for the ranking task: The goal for the ranking task is to predict whether a user will show interest in a recommended game (yes or no). This will be based on the combination of data from their current library, preferred genres, and highly rated games that the user has already played. This ranking is crucial because it allows filtering out the games that are most likely to capture the user's interest, aligning with the business goal of increasing retention and engagement on the platform. This ranking helps to clearly identify which games should be prioritized in suggestions, maximizing the impact on business decision making on which games to promote.

### Improving User Experience

By integrating this recommendation system, MetaGames aims to enrich the browsing experience on its platform. By offering more precise and relevant suggestions, it is expected to make it easier for players to discover new titles aligned with their tastes, which will increase overall satisfaction and time spent on the platform.

## Expected Impact:

- Increased User Retention: Providing more relevant and personalized recommendations should increase the time users spend on the platform and improve their engagement, which can ultimately increase retention and reduce churn.

- Increased Sales and Engagement: An effective recommendation system can boost sales as users will be more inclined to purchase games that align with their interests. Additionally, improved user experience can encourage greater engagement with the platform and its services.

- Competitive Positioning: MetaGames could differentiate itself from its competitors by offering an advanced and customized feature. This will not only attract more players but also strengthen its position in the competitive online gaming market.
