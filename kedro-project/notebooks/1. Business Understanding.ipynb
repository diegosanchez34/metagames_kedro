# Fase 1: Business Understanding

MetaGames es una plataforma de videojuegos en línea desarrollada por una empresa chilena con el objetivo de ofrecer una alternativa local a gigantes internacionales como Steam y Epic. La plataforma busca destacarse mediante precios más accesibles y sistemas de pago adaptados a las necesidades del mercado chileno. Actualmente, MetaGames proporciona a los usuarios un extenso catálogo de juegos para PC organizados en diversas categorías, así como tablas que destacan los juegos con mejores ratings y los más jugados. Con el crecimiento continuo de la plataforma, la empresa se enfoca en mejorar la experiencia del usuario a través de la implementación de nuevas características, incluida la creación de un sistema de recomendaciones personalizadas.

## Objetivos Principales:

### Desarrollar un Sistema de Recomendaciones Personalizadas

- Target para la tarea de regresión: En el contexto de este proyecto, el objetivo para la tarea de regresión es predecir la puntuación de afinidad de un juego para un usuario. Esta puntuación reflejará cuán relevante podría ser un juego nuevo para el usuario basándose en su historial de compras y los juegos en su biblioteca. Se eligió este target porque proporciona un valor cuantitativo que puede ser utilizado para ordenar las recomendaciones. Esta predicción permite que el sistema ajuste con precisión las sugerencias, lo que impacta directamente en el objetivo del negocio de mejorar la experiencia del usuario, ya que una mayor afinidad aumenta la probabilidad de interacción y compra de juegos recomendados. La selección de este target es clave para tomar decisiones posteriores, como la optimización del algoritmo de recomendaciones y la personalización de ofertas.

- Target para la tarea de clasificación: El objetivo para la tarea de clasificación es predecir si un usuario mostrará interés en un juego recomendado (sí o no). Esto se basará en la combinación de datos de su biblioteca actual, géneros preferidos y juegos con puntuaciones altas que el usuario ya ha jugado. Esta clasificación es crucial porque permite filtrar los juegos más probables de captar el interés del usuario, alineándose con el objetivo del negocio de aumentar la retención y el engagement en la plataforma. Esta clasificación ayuda a identificar claramente cuáles juegos deben ser prioritarios en las sugerencias, maximizando el impacto en la toma de decisiones empresariales sobre qué juegos promocionar.

### Mejorar la Experiencia del Usuario

Al integrar este sistema de recomendaciones, MetaGames busca enriquecer la experiencia de navegación en su plataforma. Al ofrecer sugerencias más precisas y relevantes, se espera facilitar a los jugadores el descubrimiento de nuevos títulos alineados con sus gustos, lo que aumentará la satisfacción general y el tiempo de permanencia en la plataforma.

## Impacto Esperado:

- Incremento en la Retención de Usuarios: Proporcionar recomendaciones más relevantes y personalizadas debería incrementar el tiempo que los usuarios pasan en la plataforma y mejorar su interacción, lo que, en última instancia, puede aumentar la retención y reducir la tasa de abandono.

- Aumento en las Ventas y el Engagement: Un sistema de recomendaciones efectivo puede impulsar el número de ventas, ya que los usuarios estarán más inclinados a comprar juegos que se alineen con sus intereses. Además, la mejora en la experiencia del usuario puede fomentar una mayor interacción con la plataforma y sus servicios.

- Posicionamiento Competitivo: MetaGames podría diferenciarse de sus competidores al ofrecer una característica avanzada y personalizada. Esto no solo atraerá a más jugadores, sino que también fortalecerá su posición en el competitivo mercado de videojuegos en línea.
