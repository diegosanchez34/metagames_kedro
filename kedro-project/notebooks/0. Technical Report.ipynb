## Context

MetaGames is an emerging Chilean platform that seeks to compete with giants such as Steam and Epic, offering an alternative with more affordable prices and payment methods adapted to the national population. Through its website, the platform allows players to explore an extensive catalog of PC video games, classified in various categories and presented with detailed information on ratings and popularity. In its continuous effort to improve the user experience, MetaGames plans to implement a new feature: a personalized video game recommendation system. This system will be developed using data provided by the company, including characteristics such as the release date, overall score, genre, and developer of the games. The goal is to offer suggestions based on the titles already present in the user's library, thus enriching their experience and optimizing navigation on the platform.

## Objective of the Report

The objective of this report is to document the development process of a personalized recommendation system for MetaGames using machine learning techniques. Through the use of the MCRISP-DM methodology, we seek to provide a systematic approach to optimize the user experience by suggesting video games based on their preferences and past behaviors.

## CRISP-DM methodology

This project follows the CRISP-DM (Cross Industry Standard Process for Data Mining) methodology, a standard approach widely used in data mining and machine learning projects. The methodology is organized into six phases: business understanding, data understanding, data preparation, modeling, evaluation, and deployment. Each of these stages will be detailed throughout the report, providing a structured framework for the development of the recommendation system.

## The data 

The success of any recommendation system depends largely on the quality and relevance of the data. In this case, the dataset provided by MetaGames includes key information about video games, such as their release date, overall rating, genre, and developer. This data allows for in-depth analysis that facilitates the identification of patterns and trends, which is essential for generating accurate and personalized recommendations.

## Dataset Contents

| Data | Description |
|-----------|-----------|
|game| Name of the game|
|link| Link to the game's website or store|
|release| Date of the game's release|
|peak_players| Maximum number of concurrent players at a given time|
|positive_reviews| Number of positive reviews received by the game|
|negative_reviews| Number of negative reviews received by the game|
|total_reviews| Total number of reviews for the game (positive + negative)|
|rating| Overall score or rating for the game|
|primary_genre| Primary genre the game belongs to (action, adventure, etc.)|
|store_genres| Additional genres associated with the game in the store|
|publisher| Company that publishes the game|
|developer| Company or team responsible for developing the game|
|detected_technologies| Technologies detected that the game uses (graphics engine, etc.)|
|store_asset_mod_time| Last time the game's content was modified in the store|
|review_percentage| Percentage of positive reviews out of total reviews|
|players_right_now| Number of players currently playing the game|
|24_hour_peak| Highest number of concurrent players in the last 24 hours|
|all_time_peak| Highest number of concurrent players ever|
|all_time_peak_date| Date when the highest number of concurrent players was reached|
